```python
import requests
from bs4 import BeautifulSoup
import json
import re

url = "http://www.cgv.co.kr/movies/?lt=1&ft=1"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')


movie_list = []
name_list = []
# class가 'box-image'인 div 안에 있는 모든 a 태그 가져오기
box_image_divs = soup.find_all('div', class_='box-image')
box_contents = soup.find_all('div', class_='box-contents')

for box_image_div, box_content in zip(box_image_divs, box_contents):
    
    a_tag = box_image_div.find('a')
    img_tag = box_image_div.find('img')
    name_tag = ','.join(img_tag['alt'][:-4].split())
    
    name_list.append(name_tag)
    
    num = re.search(r'\d+', a_tag['href']).group()
    movie_list.append(int(num))

print(len(name_list), len(movie_list))
```

```python
import pandas as pd
import json
review_list = []

url = 'http://www.cgv.co.kr/common/ajax/point.aspx/GetMoviePointVariableList'
headers={"User-Agent" : "Mozilla/5.0"} 
num = 10

for i in movie_list:
    temp = []
    for j in range(1,21):
        payload = { 'movieIdx': i, 'pageIndex': j, 'pageSize': num, 'orderType': 0, 'filterType': 1, 'isTotalCount' : False, 'isMyPoint' : 'false' }

        try:
            response =requests.post(url, json=payload, headers=headers)
            data = json.loads(response.text)
            
            for k in range(num):
                review = json.loads(data['d'])['List'][k]['CommentText']
                temp.append(review)
        except:
            break
        
    review_list.append(temp)
```

```python
num = len(review_list)

for i in range(num):
    print(len(review_list[i]))
```

```python
df = pd.DataFrame()
df['name'] = name_list
df['movieId'] = movie_list
df['review'] = review_list 
```

```python
df.to_csv('cgv.csv')
```
